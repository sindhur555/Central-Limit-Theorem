# A Visualization of the Central Limit Theorem

The De Moivre - Laplace Theorem states that the binomial distribution can approximate a Normal distribution with large sample size. The sample mean of a large samples(size $n$) drawn from a Binomial distribution (mean $\mu$ and standard deviation $\sigma$) approximates a Normal distribution with the same mean as the Binomial distribution and a standard deviation of $\sigma/\sqrt(n)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd
from random import randint
from matplotlib.animation import FuncAnimation
import sys
import ffmpeg
```
